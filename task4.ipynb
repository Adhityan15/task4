```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("new_retail_data.csv")
#reading data set
```

```python
pd.DataFrame(df)
```

```python
print(df.head())
```

```python
#summary of data set
print(df.info())

```

```python
print(df.describe())


```

```python
#histogram to understand distribution
df.hist(bins=30,figsize=(20,15))
plt.suptitle("histogram")
plt.show()
```

```python
#box plot to identify outliers
plt.figure(figsize=(20,15))
for i,coloumn in enumerate(df.columns,1):
    plt.subplot(5,5,i)
    sns.boxplot(y=df[coloumn])
    plt.title(f'Box plot of{coloumn}')
plt.tight_layout()
plt.show()
```

```python
#pair plot for detailed correlation check
sns.pairplot(df)
plt.suptitle("pair plot of variables",y=1.02)
plt.show()
```
